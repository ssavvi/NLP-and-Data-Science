## Balanced DataSet:  sampling with replacement
(oversampling)

```python
from itertools import count
import random
import numpy as np

# labels = [
# "Bacterial Physiological Phenomena", 
# "Biofilms",
# "Catabolite Repression", 
# "Drug Resistance, Microbial", 
# "Germ-Free Life",
# "Hemadsorption",
# "Host Microbial Interactions",
# "Host-Pathogen Interactions",
# "Microbial Interactions",
# "Microbial Viability",
# "Microbiota",
# "Nitrogen Fixation",
# "Toxin-Antitoxin Systems",
# "Virus Physiological Phenomena",
# "Virulence"
#  ]

labels = [
"Physical Phenomena",
"Chemical Phenomena",
"Metabolism",
"Cell Physiological Phenomena",
"Genetic Phenomena",
"Microbiological Phenomena",
"Physiological Phenomena",
"Reproductive and Urinary Physiological Phenomena",
"Circulatory and Respiratory Physiological Phenomena",
"Digestive System and Oral Physiological Phenomena",
"Musculoskeletal and Neural Physiological Phenomena",
"Immune System Phenomena",
"Integumentary System Physiological Phenomena",
"Ocular Physiological Phenomena",
"Plant Physiological Phenomena",
"Biological Phenomena",
"Mathematical Concepts"
 ]



retmax = 1000
minimum_number_articles = 1000
#max_loops = 3
grouped_articles = {}
# get articles
for i in range(15):
    all_articles = []
    for j in count():
        id_list = Entrez.read(Entrez.esearch(
            db='pubmed', term=labels[i], field='MESH', retmax=retmax,
            retstart=j*retmax
        ))['IdList']
        all_articles += id_list
        if len(id_list) < retmax:
            break
        ## take out at home:
#         if j >= max_loops:
#              break
        ##
    #size = max(0.1 * len(all_articles), minimum_number_articles)  # this does not make balanced
    size = minimum_number_articles  # this makes balanced, if resampling is used OR there are enough documents
    # if not resampling:
#     resample = size <= len(all_articles)
    size = min(size, len(all_articles))
    resample = False
    # if resampling:
#     resample = True
    all_articles = np.random.choice(
        all_articles,
        size=size,
        replace=resample
    )
    if len(all_articles) < minimum_number_articles:
        extra_sample = np.random.choice(
            all_articles,
            size=minimum_number_articles-len(all_articles),
            replace=True
        )
        all_articles = np.concatenate([all_articles, extra_sample])
#     all_articles = random.choices(
#         all_articles
#         , k=max(
#             int(0.1*len(all_articles), minimum_number_articles
#                ))

    grouped_articles[labels[i]] = all_articles
```

```python

import numpy as np

# id_list = grouped_articles['Catabolite Repression']
def get_unique_ids(id_list):
    article_ids_collected = []
    suffix_marker = '_'
    for article_id, article_count in zip(*np.unique(id_list,return_counts=True)):
        for i in range(article_count):
            article_ids_collected.append( str(article_id) + suffix_marker + str(i))
    return article_ids_collected

get_unique_ids(grouped_articles['Catabolite Repression'])
```

## Getting the abstracts

```python
ids = df_new.PMID.tolist()

list_of_articles = []
# iterate through Entrez queries
# ....
chunksize = 1000
#chunksize = 5
for i in range(0, len(ids), chunksize):
    all_articles_soup = BeautifulSoup(
        Entrez.efetch(
            db='pubmed', id=','.join(ids[i:i+chunksize]), retmode='xml')
    )
    articles_split = all_articles_soup.find_all('pubmedarticle')
    articles = []
    for soup in articles_split:
        abstract = soup.find('abstract')
        if abstract:
            texts = [
                x.get_text(separator='')
                for x in abstract.find_all('abstracttext')
            ]
            articles.append('\n'.join(texts))
        else:
            articles.append('')
#     articles = [x.get_text() for x in articles_split]
    list_of_articles += articles
#     break  # ends it after one loop
# ...
df_new['abstracts'] = list_of_articles
df_new.head
```
